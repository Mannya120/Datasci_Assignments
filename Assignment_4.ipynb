```python
import pandas as pd

patient_df=pd.read_csv(r"C:\Users\hp\Desktop\Data Sci\Assignments Data\Patient_Data.csv")
billing_df=pd.read_csv(r"C:\Users\hp\Desktop\Data Sci\Assignments Data\Billing_Data.csv")

patient_df.info()
billing_df.info()
patient_df.describe()

patient_df.columns

billing_column=['PatientID','Department','Doctor','BillAmount']
patient_bill=patient_df[billing_column]
print(patient_bill)

drop_col=['ReceptionistID','CheckInTime']
patient_df=patient_df.drop(columns=drop_col,errors="ignore")
print(patient_df)

dept_revenue=patient_bill.groupby("Department")["BillAmount"].sum()
print(dept_revenue)

patient_df=patient_df.drop_duplicates(subset="PatientID")
print(patient_df.shape)

mean_bill=patient_df['BillAmount'].mean()
print(mean_bill)

patient_df['BillAmount']=patient_df['BillAmount'].fillna(mean_bill)
print(patient_df)

merge_df=pd.merge(
    patient_df,
    billing_df,
    on="PatientID",
    how="inner")
print(merge_df)

new_patient=pd.DataFrame({
    "PatientID" : [106,107],
    "Name" :['Radhe','Sham'],
    "Department" :['Neurology','Orthopedics'],
    "Doctor" :['Dr. Smith','Dr. Manu'],
    "BillAmount" :[2000,3500]})
final_df=pd.concat([merge_df,new_patient],ignore_index=True)

final_df.loc[final_df["PatientID"]==106,"InsuranceCovered"]=500
final_df.loc[final_df["PatientID"]==107,"InsuranceCovered"]=1200

final_df["FinalAmount"]=(final_df["BillAmount"]-final_df["InsuranceCovered"])

print(final_df)

'''
Precision:
Precision measures how many of the predicted positive cases are actually positive.

Precision=
TP+FP
TP

TP (True Positives): Correctly predicted positives
FP (False Positives): Incorrectly predicted positives

Recall:
Recall measures how many actual positive cases were correctly identified.

Recall=
TP+FN
TP

FN (False Negatives): Actual positives missed by the model

| Metric    | Focuses On      | Important When                                                |
| --------- | --------------- | ------------------------------------------------------------- |
| Precision | False Positives | Cost of false alarm is high (e.g., fraud flagging)            |
| Recall    | False Negatives | Missing a positive case is dangerous (e.g., cancer detection) |

2.What is Cross-Validation?
Cross-validation is a model evaluation technique used to check how well your model generalizes to unseen data.
The most common type is K-Fold Cross-Validation.

How K-Fold Works:
Split dataset into K equal parts (folds).
Train model on K−1 folds.
Test on the remaining fold.
Repeat K times (each fold used once as test set).
Average the results.
Example: In 5-fold cross-validation, the model trains 5 times and produces 5 accuracy (or precision/recall) scores.

Why is it Important in Binary Classification?
a.Prevents overfitting
Ensures the model performs well on different subsets of data.

b.More reliable performance estimate
Instead of a single train-test split, you get an average performance.

c.Better for small datasets
All data gets used for both training and testing.

d.Handles class imbalance better
Using Stratified K-Fold, each fold maintains the same class ratio
'''
```
