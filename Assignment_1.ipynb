```python
n=int(input("Enter a number :"))

count = 0
for i in range(1, n+1):
    if n%i ==0:
        count += 1
        
if count == 2:
    print("Prime number")
else:
    print("Not a Prime number")
    

import random

a= random.randint(1,10)
b= random.randint(1,10)

answer=int(input(f"What is the product of {a} and {b} ? :"))

if answer == a*b:
    print("Correct !")
    
else:
    print("Incorrect !")
    print("The Product of {a} and {b} is :", a*b)

for num in range (100,201):
    
    if num%2==0:
        print(f"Even : {num} squared = {num**2}")
    else:
        print(f"Odd : {num} squared = {num**2}")
        


days=int(input("Enter number of days stayed :"))
tests=int(input("Enter the number of tests done :"))
medicine_cost=int(input("Enter the cost of medicine :"))

room_bill=days*2000
tests_bill=tests*300
doctor_fee=1500

total_bill= room_bill + tests_bill + doctor_fee + medicine_cost

if total_bill > 10000:
    discount= total_bill*0.1
    
else:
    discount=0
    

print("\n-------HOSPITAL BILL SUMMARY-------")
print("Doctor Consultation Fee :",doctor_fee)
print("Total Room Charges :",room_bill)
print("Lab Test Bill :",tests_bill)
print("Total Medicine Cost :",medicine_cost)
print("Total Before Discount :",total_bill)
print("Discounted Amount:",discount)
print("Final Bill :",total_bill-discount)
```
