```python
import pandas as pd

df=pd.read_csv(r"C:\Users\hp\Desktop\Data Sci\Assignments Data\sales_data_with_discounts.csv")

df.info()

df.describe()

df.columns

df.head()

num_col=df.select_dtypes(include=["int64","float64"]).columns
cat_col=df.select_dtypes(include=["object"]).columns

print("\n-----NUMERICAL D-TYPE-----")
print("Numerical Dtype :",num_col)

print("\n-----CATEGORICAL D-TYPE-----")
print("Categorical Dtype :",cat_col)

df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
       'Discount Amount', 'Net Sales Value']].mean()

df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
       'Discount Amount', 'Net Sales Value']].median()

df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
       'Discount Amount', 'Net Sales Value']].mode()

df[['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
       'Discount Amount', 'Net Sales Value']].std()

stats=pd.DataFrame({
    "Mean":df[num_col].mean(),
    "Median" :df[num_col].median(),
    "Mode" :df[num_col].mode().iloc[0],
    "Std Deviation" :df[num_col].std()
    })
print(stats)

import matplotlib.pyplot as plt

for col in num_col:
    plt.figure()
    plt.hist(df[col])
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

for col in num_col:
    plt.figure()
    plt.boxplot(df[col])
    plt.title(f"Box Plot of {col}")
    plt.ylabel(col)
    plt.show()
    
for col in cat_col:
    plt.figure(figsize=(8,4))
    df[col].value_counts().plot(kind="bar")
    plt.title(f"Categorical Distribution : {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()

```
